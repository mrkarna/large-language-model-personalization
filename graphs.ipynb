##Plotting the results obtained.

Results for our model (MAE, RMSE) for k=1,2.

```python
import matplotlib.pyplot as plt

k_values = [1, 2]
mae_values = [1.046, 1.053]
rmse_values = [0.6848, 0.7004]

plt.subplot(1, 2, 1)
plt.plot(k_values, mae_values, marker='o', linestyle='-', color='blue')
plt.title('MAE vs. k')
plt.xlabel('k')
plt.ylabel('MAE')
plt.xticks(k_values)

plt.subplot(1, 2, 2)
plt.plot(k_values, rmse_values, marker='o', linestyle='-', color='green')
plt.title('RMSE vs. k')
plt.xlabel('k')
plt.ylabel('RMSE')
plt.xticks(k_values)

plt.tight_layout()

plt.show()
```

Comparison of baseline and ourModel (MAE, RMSE) for k=1

```python
import matplotlib.pyplot as plt

models = ['baseline', 'ourModel']
mae_values = [1.0478, 1.04646]
rmse_values = [0.72, 0.6848]

plt.subplot(1, 2, 1)
plt.plot(models, mae_values, marker='o', linestyle='-', color='blue')
plt.title('MAE Comparison')
plt.xlabel('k=1')
plt.ylabel('MAE')

plt.scatter(models, mae_values, color='red', marker='x', label='Points')
for i, txt in enumerate(["baseline", "ourModel"]):
    plt.annotate(txt, (models[i], mae_values[i]), textcoords="offset points", xytext=(0,10), ha='center', fontsize=8)

plt.subplot(1, 2, 2)
plt.plot(models, rmse_values, marker='o', linestyle='-', color='purple')
plt.title('RMSE Comparison')
plt.xlabel('k=1')
plt.ylabel('RMSE')

plt.scatter(models, rmse_values, color='red', marker='x', label='Points')
for i, txt in enumerate(["baseline", "ourModel"]):
    plt.annotate(txt, (models[i], rmse_values[i]), textcoords="offset points", xytext=(0,10), ha='center', fontsize=8)

plt.tight_layout()

plt.show()
```
